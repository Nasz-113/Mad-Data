```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from imblearn.combine import SMOTEENN
from imblearn.pipeline import Pipeline

# Load dataset
nadata = pd.read_csv("merged_data.csv")

# Split features and target
Xna = nadata.drop("churned_30d", axis=1)
yna = nadata["churned_30d"]

# Identify categorical & numerical columns
categorical_cols = Xna.select_dtypes(include=["object", "category"]).columns.tolist()
numeric_cols = Xna.select_dtypes(include=["int64", "float64"]).columns.tolist()

print("Categorical columns:", categorical_cols)
print("Numeric columns:", numeric_cols)

# Preprocessing
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)

# Logistic Regression model
log_reg = LogisticRegression(
    max_iter=1000,
    solver="saga",
    class_weight=None
)

# Pipeline with SMOTEENN
pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("smoteenn", SMOTEENN(random_state=42)),
    ("model", log_reg)
])

# Cross-validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Get cross-validated probabilities
y_pred_proba = cross_val_predict(pipeline, Xna, yna, cv=cv, method='predict_proba')

# Apply custom threshold 0.3
threshold = 0.6
y_pred = (y_pred_proba[:, 1] >= threshold).astype(int)

# Compute F1 score
f1 = f1_score(yna, y_pred)
print("Overall F1 score:", f1)

# Confusion matrix
cm = confusion_matrix(yna, y_pred)
print("\nConfusion Matrix:\n", cm)

# Detailed metrics
print("\nClassification Report:\n", classification_report(yna, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
# Decision Tree model
dtree = DecisionTreeClassifier(
    max_depth=None,        # or set a value to prevent overfitting
    class_weight= None, #"balanced",  # handle imbalance
    random_state=42
)

# Pipeline with SMOTEENN
pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("smoteenn", SMOTEENN(random_state=42)),
    ("model", dtree)
])

# Cross-validation strategy
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Get cross-validated predictions
y_pred = cross_val_predict(pipeline, Xna, yna, cv=cv)

# Compute F1 score
f1 = f1_score(yna, y_pred)
print("Overall F1 score:", f1)

# Confusion matrix
cm = confusion_matrix(yna, y_pred)
print("\nConfusion Matrix:\n", cm)

# Detailed metrics
print("\nClassification Report:\n", classification_report(yna, y_pred))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(
    use_label_encoder=False,
    eval_metric="auc",   # prevent warnings
    random_state=42,
    scale_pos_weight=1       # keep default since SMOTE balances data
)

pipeline_xgb = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("smote", SMOTE(random_state=42)),
    ("model", xgb)
])

y_pred_xgb = cross_val_predict(pipeline_xgb, Xna, yna, cv=cv)

print("\n=== XGBoost + SMOTE ===")
print("F1 Score:", f1_score(yna, y_pred_xgb))
print("Confusion Matrix:\n", confusion_matrix(yna, y_pred_xgb))
print("\nClassification Report:\n", classification_report(yna, y_pred_xgb))
```
